```python
F_I=np.fft.fft2(I)
Fshift_I=np.fft.fftshift(F_I)

M,N=F_I.shape
H=np.zeros((M,N),dtype=np.float32)
D0=50
for u in range(M):
    for v in range(N):
        D=np.sqrt((u-M/2)**2+(v-N/2)**2)
        if D<=D0:
            H[u,v]=1
        else:
            H[u,v]=0
Gshift=Fshift_I*H
G=np.fft.ifftshift(Gshift)
g=np.abs(np.fft.ifft2(G))
plt.figure()
plt.subplot(121)
plt.imshow(np.log1p(np.abs(F_I)),cmap='gray')
plt.subplot(122)
plt.imshow(np.log1p(np.abs(Fshift_I)),cmap='gray')
plt.figure()

plt.subplot(221)

plt.imshow(H,cmap='gray')
plt.subplot(222)

plt.imshow(np.log1p(np.abs(Gshift)),cmap='gray')
plt.figure()

plt.imshow(g,cmap='gray')






```

```python
M,N=F_I.shape
H3=np.zeros((M,N),dtype=np.float32)
D0=50
for u in range(M):
    for v in range(N):
        D=np.sqrt((u-M/2)**2+(v-N/2)**2)
        H3[u,v]=np.exp(-D**2/(2*D0*D0))
Gshift=Fshift_I*H3
G=np.fft.ifftshift(Gshift)
g3=np.abs(np.fft.ifft2(G))
plt.figure(figsize=(8,8))
plt.subplot(2,2,1)
plt.imshow(H3,cmap='gray')
plt.subplot(2,2,2)
plt.imshow(g3,cmap='gray')
```
